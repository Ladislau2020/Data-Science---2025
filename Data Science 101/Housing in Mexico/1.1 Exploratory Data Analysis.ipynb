**Task 1.3.4:** Clean df1 by dropping rows with NaN values. Then remove the "$" and "," characters from "price_usd" and recast the values in the column as floats. 

```python
# Drop null values from df1
df1 = df1.dropna()
df1.info()
```

```python
# Clean "price_usd" column in df1
df1['price_usd'] = (
    df1['price_usd']
    .str.replace('$', '', regex=False)
    .str.replace(',', '')
    .astype(float))
df1.info()
```
